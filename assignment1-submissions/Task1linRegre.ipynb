```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("C:\\Tech clubs\\ACMW\\mpgLinearRegression.csv")
df
```

```python
import math

# Step 1: Replace '?' and 'nan' (strings) with NaN
df['horsepower'] = df['horsepower'].replace(['?', 'nan'], pd.NA)

# Step 2: Convert to numeric (this will coerce invalid entries to NaN)
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

# Step 3: Calculate the floored median and replace NaN values with it
floored_median = math.floor(df['horsepower'].median())
df['horsepower'] = df['horsepower'].fillna(floored_median)

df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin']], df.mpg)
```

```python
reg.coef_
```

```python
reg.intercept_ # this is the constant at the end
```
